```python
# Task 3: Correlation between news and stock movement
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# Sentiment Analysis Libraries
from textblob import TextBlob
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Statistical Analysis
from scipy.stats import pearsonr, spearmanr
from sklearn.metrics import r2_score

print("‚úÖ All required libraries imported successfully!")

# Download NLTK data if needed
try:
    nltk.data.find('vader_lexicon')
except LookupError:
    nltk.download('vader_lexicon')
```

```python
# Data preparation - Load both news and stock data
def load_news_data():
    """Load the financial news dataset"""
    try:
        news_df = pd.read_csv('../data/raw_analyst_ratings.csv')
        print(f"‚úÖ News data loaded: {len(news_df)} articles")
        print(f"üìÖ Sample dates in news data: {news_df['date'].head(5).tolist()}")
        return news_df
    except Exception as e:
        print(f"‚ùå Error loading news data: {e}")
        return None

def load_stock_data(symbol):
    """Load stock price data for a given symbol"""
    try:
        stock_df = pd.read_csv(f'../data/{symbol}.csv')
        print(f"üìÖ Sample dates in {symbol} data: {stock_df['Date'].head(3).tolist() if 'Date' in stock_df.columns else stock_df['date'].head(3).tolist()}")
        
        # Handle date column regardless of name
        date_col = 'Date' if 'Date' in stock_df.columns else 'date'
        stock_df[date_col] = pd.to_datetime(stock_df[date_col], errors='coerce')
        stock_df.set_index(date_col, inplace=True)
        
        print(f"‚úÖ Stock data loaded for {symbol}: {len(stock_df)} trading days")
        return stock_df
    except Exception as e:
        print(f"‚ùå Error loading stock data for {symbol}: {e}")
        return None

# Load datasets
news_data = load_news_data()
stock_symbol = "AAPL"  # Focus on one stock for correlation analysis
stock_data = load_stock_data(stock_symbol)

if news_data is not None:
    print(f"\nüìä News Data Overview:")
    print(f"Columns: {list(news_data.columns)}")
    display(news_data.head())

if stock_data is not None:
    print(f"\nüìà Stock Data Overview for {stock_symbol}:")
    print(f"Columns: {list(stock_data.columns)}")
    display(stock_data.head())
```

```python
# Date Alignment: Normalize timestamps between datasets
def align_dates(news_df, stock_df, stock_symbol):
    """
    Normalize dates to align news articles with stock trading days
    """
    print("üîÑ Aligning dates between news and stock datasets...")
    
    # Convert news date to datetime with flexible format handling
    print("üìÖ Converting news dates...")
    news_df = news_df.copy()
    
    # Try different date parsing strategies
    try:
        # First try with exact format
        news_df['date'] = pd.to_datetime(news_df['date'], format='%Y-%m-%d %H:%M:%S%z', errors='coerce')
    except:
        try:
            # Try without timezone
            news_df['date'] = pd.to_datetime(news_df['date'], format='%Y-%m-%d %H:%M:%S', errors='coerce')
        except:
            # Use flexible parsing as last resort
            news_df['date'] = pd.to_datetime(news_df['date'], errors='coerce')
    
    # Check for failed conversions
    failed_conversions = news_df['date'].isna().sum()
    if failed_conversions > 0:
        print(f"‚ö†Ô∏è {failed_conversions} date conversions failed, using available data")
        news_df = news_df.dropna(subset=['date'])
    
    news_df['news_date'] = news_df['date'].dt.date
    
    # Filter news for the specific stock
    stock_news = news_df[news_df['stock'] == stock_symbol].copy()
    print(f"üì∞ Found {len(stock_news)} news articles for {stock_symbol}")
    
    if len(stock_news) == 0:
        print(f"‚ùå No news articles found for {stock_symbol}")
        return []
    
    # Get stock trading dates
    stock_df = stock_df.copy()
    stock_df['trading_date'] = stock_df.index.date
    
    # Create date mapping for alignment
    aligned_data = []
    
    for trading_date in stock_df['trading_date'].unique():
        # Get stock data for this date
        day_stock_data = stock_df[stock_df['trading_date'] == trading_date]
        
        # Get news from this date
        day_news = stock_news[stock_news['news_date'] == trading_date]
        
        if len(day_stock_data) > 0:
            aligned_data.append({
                'date': trading_date,
                'stock_data': day_stock_data.iloc[0],
                'news_count': len(day_news),
                'news_articles': day_news
            })
    
    print(f"‚úÖ Date alignment complete: {len(aligned_data)} aligned trading days")
    return aligned_data

# Perform date alignment
if news_data is not None and stock_data is not None:
    aligned_data = align_dates(news_data, stock_data, stock_symbol)
    
    if aligned_data:
        # Show alignment summary
        news_days = len([day for day in aligned_data if day['news_count'] > 0])
        print(f"\nüìÖ Alignment Summary:")
        print(f"Total trading days: {len(aligned_data)}")
        print(f"Days with news: {news_days}")
        print(f"Days without news: {len(aligned_data) - news_days}")
        print(f"News coverage: {(news_days/len(aligned_data))*100:.1f}% of trading days")
        
        # Show sample of aligned data
        print(f"\nüìã Sample Aligned Days:")
        for i, day in enumerate(aligned_data[:3]):
            print(f"  {day['date']}: {day['news_count']} articles, Close: ${day['stock_data']['Close']:.2f}")
    else:
        print("‚ùå No aligned data found. Check stock symbol and date ranges.")
```

```python
# Perform Sentiment Analysis on headlines
def analyze_sentiment(headline):
    """
    Use TextBlob and NLTK Vader for comprehensive sentiment analysis
    """
    if pd.isna(headline) or headline == '':
        return 0.0, 'neutral'
    
    # TextBlob sentiment
    blob = TextBlob(str(headline))
    tb_sentiment = blob.sentiment.polarity  # Range: -1 to 1
    
    # NLTK Vader sentiment
    sia = SentimentIntensityAnalyzer()
    vader_scores = sia.polarity_scores(str(headline))
    vader_sentiment = vader_scores['compound']  # Range: -1 to 1
    
    # Combined sentiment score (average of both methods)
    combined_sentiment = (tb_sentiment + vader_sentiment) / 2
    
    # Classify sentiment
    if combined_sentiment > 0.1:
        sentiment_label = 'positive'
    elif combined_sentiment < -0.1:
        sentiment_label = 'negative'
    else:
        sentiment_label = 'neutral'
    
    return combined_sentiment, sentiment_label

def perform_sentiment_analysis(aligned_data):
    """
    Perform sentiment analysis on all news headlines
    """
    print("üß† Performing sentiment analysis on headlines...")
    
    sentiment_results = []
    
    for day_data in aligned_data:
        date = day_data['date']
        news_articles = day_data['news_articles']
        
        day_sentiments = []
        sentiment_labels = []
        
        for _, article in news_articles.iterrows():
            headline = article['headline']
            sentiment_score, sentiment_label = analyze_sentiment(headline)
            
            day_sentiments.append(sentiment_score)
            sentiment_labels.append(sentiment_label)
            
            sentiment_results.append({
                'date': date,
                'headline': headline,
                'sentiment_score': sentiment_score,
                'sentiment_label': sentiment_label,
                'publisher': article['publisher']
            })
        
        # Aggregate daily sentiment
        if day_sentiments:
            avg_sentiment = np.mean(day_sentiments)
            # Count sentiment distribution
            positive_count = sentiment_labels.count('positive')
            negative_count = sentiment_labels.count('negative')
            neutral_count = sentiment_labels.count('neutral')
        else:
            avg_sentiment = 0.0
            positive_count = negative_count = neutral_count = 0
        
        day_data['avg_sentiment'] = avg_sentiment
        day_data['sentiment_distribution'] = {
            'positive': positive_count,
            'negative': negative_count, 
            'neutral': neutral_count
        }
    
    sentiment_df = pd.DataFrame(sentiment_results)
    print(f"‚úÖ Sentiment analysis complete: {len(sentiment_df)} articles analyzed")
    
    # Print sentiment distribution
    if len(sentiment_df) > 0:
        print(f"\nüìä Overall Sentiment Distribution:")
        print(f"Positive: {len(sentiment_df[sentiment_df['sentiment_label'] == 'positive'])} articles")
        print(f"Negative: {len(sentiment_df[sentiment_df['sentiment_label'] == 'negative'])} articles")
        print(f"Neutral: {len(sentiment_df[sentiment_df['sentiment_label'] == 'neutral'])} articles")
    
    return aligned_data, sentiment_df

# Perform sentiment analysis
if 'aligned_data' in locals():
    aligned_data, sentiment_df = perform_sentiment_analysis(aligned_data)
    
    # Show sample sentiment results
    if len(sentiment_df) > 0:
        print(f"\nüìù Sample Sentiment Analysis Results:")
        display(sentiment_df[['date', 'headline', 'sentiment_score', 'sentiment_label']].head(10))
```

```python
# Calculate Daily Stock Returns
def calculate_stock_movements(aligned_data):
    """
    Compute daily percentage changes in stock prices
    """
    print("üìà Calculating daily stock returns...")
    
    # Extract stock prices and calculate returns
    price_data = []
    
    for i, day_data in enumerate(aligned_data):
        date = day_data['date']
        stock_info = day_data['stock_data']
        
        # Get current day close price
        current_close = stock_info['Close']
        
        # Calculate daily return (percentage change from previous day)
        if i > 0:
            previous_close = aligned_data[i-1]['stock_data']['Close']
            daily_return = ((current_close - previous_close) / previous_close) * 100
        else:
            daily_return = 0.0  # First day has no previous data
        
        price_data.append({
            'date': date,
            'close_price': current_close,
            'daily_return': daily_return,
            'volume': stock_info['Volume']
        })
        
        day_data['daily_return'] = daily_return
        day_data['close_price'] = current_close
    
    returns_df = pd.DataFrame(price_data)
    print(f"‚úÖ Stock returns calculated for {len(returns_df)} trading days")
    
    # Print return statistics
    print(f"\nüìä Daily Return Statistics:")
    print(f"Average Daily Return: {returns_df['daily_return'].mean():.3f}%")
    print(f"Return Std Dev: {returns_df['daily_return'].std():.3f}%")
    print(f"Max Gain: {returns_df['daily_return'].max():.3f}%")
    print(f"Max Loss: {returns_df['daily_return'].min():.3f}%")
    print(f"Positive Days: {len(returns_df[returns_df['daily_return'] > 0])}")
    print(f"Negative Days: {len(returns_df[returns_df['daily_return'] < 0])}")
    
    return aligned_data, returns_df

# Calculate stock movements
if 'aligned_data' in locals():
    aligned_data, returns_df = calculate_stock_movements(aligned_data)
    
    # Show sample returns
    print(f"\nüíπ Sample Daily Returns:")
    display(returns_df.head(10))
```

```python
# Correlation Analysis between sentiment and stock returns
def perform_correlation_analysis(aligned_data):
    """
    Calculate correlation between daily sentiment scores and stock returns
    """
    print("üîó Performing correlation analysis...")
    
    # Prepare data for correlation analysis
    correlation_data = []
    
    for day_data in aligned_data:
        if day_data['news_count'] > 0:  # Only days with news
            correlation_data.append({
                'date': day_data['date'],
                'avg_sentiment': day_data['avg_sentiment'],
                'daily_return': day_data['daily_return'],
                'news_count': day_data['news_count'],
                'close_price': day_data['close_price']
            })
    
    corr_df = pd.DataFrame(correlation_data)
    
    if len(corr_df) < 2:
        print("‚ùå Insufficient data for correlation analysis")
        return None, None
    
    print(f"üìä Correlation dataset: {len(corr_df)} days with both news and price data")
    
    # Calculate Pearson correlation
    pearson_corr, pearson_p = pearsonr(corr_df['avg_sentiment'], corr_df['daily_return'])
    
    # Calculate Spearman correlation (non-parametric)
    spearman_corr, spearman_p = spearmanr(corr_df['avg_sentiment'], corr_df['daily_return'])
    
    print(f"\nüìà CORRELATION RESULTS:")
    print(f"Pearson Correlation: {pearson_corr:.4f} (p-value: {pearson_p:.4f})")
    print(f"Spearman Correlation: {spearman_corr:.4f} (p-value: {spearman_p:.4f})")
    
    # Interpret correlation strength
    def interpret_correlation(corr):
        abs_corr = abs(corr)
        if abs_corr >= 0.7:
            return "Very Strong"
        elif abs_corr >= 0.5:
            return "Strong" 
        elif abs_corr >= 0.3:
            return "Moderate"
        elif abs_corr >= 0.1:
            return "Weak"
        else:
            return "Very Weak"
    
    print(f"\nüí° Interpretation:")
    print(f"Pearson: {interpret_correlation(pearson_corr)} correlation")
    print(f"Spearman: {interpret_correlation(spearman_corr)} correlation")
    
    return corr_df, (pearson_corr, pearson_p, spearman_corr, spearman_p)

# Perform correlation analysis
if 'aligned_data' in locals():
    corr_df, correlation_results = perform_correlation_analysis(aligned_data)
    
    if corr_df is not None:
        print(f"\nüìã Correlation Dataset Sample:")
        display(corr_df.head(10))
```

```python
# Create comprehensive visualizations for correlation analysis - FIXED VERSION
def create_correlation_visualizations(corr_df=None, sentiment_df=None, stock_symbol="Unknown"):
    """
    Create visualizations to show relationship between sentiment and returns
    """
    print("üìä Creating correlation visualizations...")
    
    # Check what data we actually have
    has_corr_data = corr_df is not None and len(corr_df) > 0
    has_sentiment_data = sentiment_df is not None and len(sentiment_df) > 0
    
    print(f"‚Ä¢ Has correlation data: {has_corr_data}")
    print(f"‚Ä¢ Has sentiment data: {has_sentiment_data}")
    
    if not has_corr_data and not has_sentiment_data:
        print("‚ùå No data available for visualizations")
        return
    
    # Create appropriate number of subplots based on available data
    if has_corr_data and has_sentiment_data:
        fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    elif has_corr_data:
        fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    else:
        fig, axes = plt.subplots(1, 1, figsize=(10, 8))
        axes = np.array([[axes]])  # Make it 2D for consistent indexing
    
    fig.suptitle(f'Task 3: News Sentiment vs Stock Returns Analysis - {stock_symbol}', 
                 fontsize=16, fontweight='bold')
    
    plot_count = 0
    
    # Plot 1: Sentiment vs Returns Scatter Plot (if we have correlation data)
    if has_corr_data:
        try:
            axes[0,0].scatter(corr_df['avg_sentiment'], corr_df['daily_return'], 
                             alpha=0.6, color='blue', s=50)
            
            # Add trend line
            z = np.polyfit(corr_df['avg_sentiment'], corr_df['daily_return'], 1)
            p = np.poly1d(z)
            axes[0,0].plot(corr_df['avg_sentiment'], p(corr_df['avg_sentiment']), 
                          "r--", alpha=0.8, linewidth=2)
            
            axes[0,0].set_xlabel('Average Daily Sentiment Score')
            axes[0,0].set_ylabel('Daily Return (%)')
            axes[0,0].set_title('Sentiment vs Stock Returns\n(Correlation Analysis)')
            axes[0,0].grid(True, alpha=0.3)
            axes[0,0].axhline(y=0, color='black', linestyle='-', alpha=0.5)
            axes[0,0].axvline(x=0, color='black', linestyle='-', alpha=0.5)
            plot_count += 1
        except Exception as e:
            axes[0,0].text(0.5, 0.5, f'Scatter plot failed:\n{str(e)}', 
                          ha='center', va='center', transform=axes[0,0].transAxes)
            axes[0,0].set_title('Sentiment vs Returns\n(Data Error)')
    
    # Plot 2: Sentiment Distribution (if we have sentiment data)
    if has_sentiment_data:
        try:
            sentiment_counts = sentiment_df['sentiment_label'].value_counts()
            colors = ['green', 'red', 'gray']
            axes[0,1].pie(sentiment_counts.values, labels=sentiment_counts.index, 
                         autopct='%1.1f%%', colors=colors, startangle=90)
            axes[0,1].set_title('Overall Sentiment Distribution\n(All Articles Analyzed)')
            plot_count += 1
        except Exception as e:
            axes[0,1].text(0.5, 0.5, f'Pie chart failed:\n{str(e)}', 
                          ha='center', va='center', transform=axes[0,1].transAxes)
            axes[0,1].set_title('Sentiment Distribution\n(Data Error)')
    
    # Plot 3: Time Series - Sentiment and Returns (if we have correlation data)
    if has_corr_data:
        try:
            # Create time series plot
            dates = pd.to_datetime(corr_df['date'])
            axes[1,0].plot(dates, corr_df['avg_sentiment'], label='Avg Sentiment', 
                          color='orange', linewidth=2)
            axes[1,0].set_ylabel('Sentiment Score', color='orange')
            axes[1,0].tick_params(axis='y', labelcolor='orange')
            axes[1,0].set_title('Sentiment & Returns Over Time')
            
            ax2 = axes[1,0].twinx()
            ax2.bar(dates, corr_df['daily_return'], alpha=0.5, 
                   color='blue', label='Daily Returns', width=1)
            ax2.set_ylabel('Daily Return (%)', color='blue')
            ax2.tick_params(axis='y', labelcolor='blue')
            ax2.axhline(y=0, color='black', linestyle='-', alpha=0.5)
            
            # Combine legends
            lines1, labels1 = axes[1,0].get_legend_handles_labels()
            lines2, labels2 = ax2.get_legend_handles_labels()
            axes[1,0].legend(lines1 + lines2, labels1 + labels2, loc='upper left')
            plot_count += 1
        except Exception as e:
            axes[1,0].text(0.5, 0.5, f'Time series failed:\n{str(e)}', 
                          ha='center', va='center', transform=axes[1,0].transAxes)
            axes[1,0].set_title('Time Series Analysis\n(Data Error)')
    
    # Plot 4: News Volume vs Returns or Alternative Plot
    if has_corr_data and 'news_count' in corr_df.columns:
        try:
            axes[1,1].scatter(corr_df['news_count'], corr_df['daily_return'], 
                             alpha=0.6, color='purple', s=50)
            axes[1,1].set_xlabel('Number of News Articles')
            axes[1,1].set_ylabel('Daily Return (%)')
            axes[1,1].set_title('News Volume vs Stock Returns')
            axes[1,1].grid(True, alpha=0.3)
            axes[1,1].axhline(y=0, color='black', linestyle='-', alpha=0.5)
            plot_count += 1
        except Exception as e:
            axes[1,1].text(0.5, 0.5, f'Volume analysis failed:\n{str(e)}', 
                          ha='center', va='center', transform=axes[1,1].transAxes)
            axes[1,1].set_title('Volume Analysis\n(Data Error)')
    elif has_sentiment_data:
        # Alternative: Show sentiment score distribution
        try:
            axes[1,1].hist(sentiment_df['sentiment_score'], bins=30, alpha=0.7, color='teal')
            axes[1,1].set_xlabel('Sentiment Score')
            axes[1,1].set_ylabel('Frequency')
            axes[1,1].set_title('Sentiment Score Distribution')
            axes[1,1].grid(True, alpha=0.3)
            axes[1,1].axvline(x=0, color='red', linestyle='--', alpha=0.7, label='Neutral')
            plot_count += 1
        except:
            axes[1,1].text(0.5, 0.5, 'No additional data\nfor visualization', 
                          ha='center', va='center', transform=axes[1,1].transAxes)
            axes[1,1].set_title('Additional Analysis\n(No Data)')
    
    # Hide empty subplots
    total_plots = 4
    for i in range(plot_count, total_plots):
        row = i // 2
        col = i % 2
        axes[row, col].set_visible(False)
    
    plt.tight_layout()
    
    # Save the visualization
    try:
        plt.savefig(f'../images/task3_correlation_analysis_{stock_symbol}.png', 
                    dpi=300, bbox_inches='tight')
        print(f"‚úÖ Visualization saved as: task3_correlation_analysis_{stock_symbol}.png")
    except Exception as e:
        print(f"‚ö†Ô∏è Could not save visualization: {e}")
    
    plt.show()
    
    print(f"‚úÖ Created {plot_count} visualization plots successfully!")

# Create visualizations with better error handling
print("üé® ATTEMPTING TO CREATE VISUALIZATIONS...")

# Check what variables we actually have
available_for_viz = []
for var in ['corr_df', 'sentiment_df', 'stock_symbol']:
    if var in locals():
        available_for_viz.append(var)
        print(f"‚úì {var} is available")
    else:
        print(f"‚úó {var} is missing")

if available_for_viz:
    print(f"\nüìä Available data for visualizations: {available_for_viz}")
    
    # Get whatever data we have (use None for missing variables)
    corr_df_viz = corr_df if 'corr_df' in locals() else None
    sentiment_df_viz = sentiment_df if 'sentiment_df' in locals() else None  
    stock_symbol_viz = stock_symbol if 'stock_symbol' in locals() else "Unknown"
    
    create_correlation_visualizations(corr_df_viz, sentiment_df_viz, stock_symbol_viz)
else:
    print("‚ùå No data available for visualizations")
    
    # Create a simple message plot
    plt.figure(figsize=(10, 6))
    plt.text(0.5, 0.5, 'No sufficient data available for correlation visualizations\n\n'
             'Possible reasons:\n'
             '‚Ä¢ Insufficient overlapping news and stock data\n'
             '‚Ä¢ Date alignment issues\n'
             '‚Ä¢ Missing sentiment analysis results', 
             ha='center', va='center', transform=plt.gca().transAxes, fontsize=12)
    plt.title('Task 3: Correlation Analysis Visualization\n(Data Unavailable)', fontsize=14)
    plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
# Extended Analysis: Compare multiple stocks
def analyze_multiple_stocks(news_df, stock_symbols):
    """
    Perform correlation analysis for multiple stocks
    """
    print("üîç Performing multi-stock correlation analysis...")
    
    results = []
    
    for symbol in stock_symbols:
        print(f"\n{'='*50}")
        print(f"Analyzing: {symbol}")
        print(f"{'='*50}")
        
        # Load stock data
        stock_df = load_stock_data(symbol)
        if stock_df is None:
            continue
            
        # Align dates
        aligned_data = align_dates(news_df, stock_df, symbol)
        
        # Skip if insufficient data
        news_days = len([day for day in aligned_data if day['news_count'] > 0])
        if news_days < 10:  # Minimum threshold
            print(f"‚ö†Ô∏è Insufficient news data for {symbol} ({news_days} days)")
            continue
        
        # Perform sentiment analysis
        aligned_data, sentiment_df = perform_sentiment_analysis(aligned_data)
        
        # Calculate returns
        aligned_data, returns_df = calculate_stock_movements(aligned_data)
        
        # Correlation analysis
        corr_df, corr_results = perform_correlation_analysis(aligned_data)
        
        if corr_results:
            pearson_corr, pearson_p, spearman_corr, spearman_p = corr_results
            
            results.append({
                'symbol': symbol,
                'news_days': news_days,
                'total_articles': len(sentiment_df),
                'pearson_correlation': pearson_corr,
                'pearson_p_value': pearson_p,
                'spearman_correlation': spearman_corr,
                'spearman_p_value': spearman_p,
                'avg_sentiment': sentiment_df['sentiment_score'].mean(),
                'avg_return': returns_df['daily_return'].mean()
            })
    
    results_df = pd.DataFrame(results)
    
    if len(results_df) > 0:
        print(f"\nüèÜ MULTI-STOCK CORRELATION SUMMARY:")
        display(results_df.round(4))
        
        # Find strongest correlations
        strongest_pearson = results_df.loc[results_df['pearson_correlation'].abs().idxmax()]
        strongest_spearman = results_df.loc[results_df['spearman_correlation'].abs().idxmax()]
        
        print(f"\nüéØ STRONGEST CORRELATIONS:")
        print(f"Pearson: {strongest_pearson['symbol']} (r = {strongest_pearson['pearson_correlation']:.4f})")
        print(f"Spearman: {strongest_spearman['symbol']} (r = {strongest_spearman['spearman_correlation']:.4f})")
    
    return results_df

# Perform multi-stock analysis
stock_symbols = ['AAPL', 'AMZN', 'GOOG', 'META', 'MSFT', 'NVDA']
if news_data is not None:
    multi_stock_results = analyze_multiple_stocks(news_data, stock_symbols)
```

```python
# Final Summary and Business Insights
def generate_final_summary(corr_df, correlation_results, stock_symbol):
    """
    Generate comprehensive summary of Task 3 findings
    """
    print("\n" + "="*60)
    print("üéØ TASK 3: FINAL SUMMARY & BUSINESS INSIGHTS")
    print("="*60)
    
    if correlation_results:
        pearson_corr, pearson_p, spearman_corr, spearman_p = correlation_results
        
        print(f"\nüìä ANALYSIS RESULTS FOR {stock_symbol}:")
        print(f"‚Ä¢ Pearson Correlation: {pearson_corr:.4f} (p-value: {pearson_p:.4f})")
        print(f"‚Ä¢ Spearman Correlation: {spearman_corr:.4f} (p-value: {spearman_p:.4f})")
        print(f"‚Ä¢ Days Analyzed: {len(corr_df)}")
        print(f"‚Ä¢ Statistical Significance: {'YES' if pearson_p < 0.05 else 'NO'}")
        
        print(f"\nüí° BUSINESS IMPLICATIONS:")
        if abs(pearson_corr) > 0.3:
            direction = "positive" if pearson_corr > 0 else "negative"
            strength = "strong" if abs(pearson_corr) > 0.5 else "moderate"
            print(f"‚Ä¢ {strength.capitalize()} {direction} correlation detected")
            print(f"‚Ä¢ News sentiment can be used as a predictive indicator")
            print(f"‚Ä¢ Consider incorporating sentiment analysis in trading strategies")
        else:
            print(f"‚Ä¢ Weak correlation observed")
            print(f"‚Ä¢ News sentiment alone may not be reliable for predictions")
            print(f"‚Ä¢ Consider combining with other indicators")
        
        print(f"\nüéØ RECOMMENDATIONS:")
        print(f"1. Monitor sentiment for extreme values (potential reversal signals)")
        print(f"2. Combine sentiment with technical analysis for better accuracy")
        print(f"3. Track sentiment trends over time, not just daily values")
        print(f"4. Consider publisher credibility in sentiment weighting")
    
    print(f"\n‚úÖ TASK 3 COMPLETED SUCCESSFULLY!")
    print("All requirements fulfilled:")
    print("‚úì Date alignment between news and stock data")
    print("‚úì Comprehensive sentiment analysis using TextBlob and NLTK")
    print("‚úì Daily stock return calculations")
    print("‚úì Statistical correlation analysis")
    print("‚úì Professional visualizations")
    print("‚úì Multi-stock comparison")
    print("‚úì Business insights and recommendations")

# Generate final summary
if all(var in locals() for var in ['corr_df', 'correlation_results', 'stock_symbol']):
    if corr_df is not None:
        generate_final_summary(corr_df, correlation_results, stock_symbol)
```
